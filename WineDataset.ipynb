```python
# Importar bibliotecas necessárias
from sklearn.datasets import load_wine
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o dataset
wine_data = load_wine()

# Converter para um DataFrame do Pandas para facilitar a análise
wine_df = pd.DataFrame(data=wine_data.data, columns=wine_data.feature_names)

# Adicionar a coluna alvo (target), que contém as classificações
wine_df['target'] = wine_data.target
```

O Wine Recognition Dataset é um conjunto de dados que faz parte dos datasets de brinquedo disponibilizados pela biblioteca scikit-learn e contém informações sobre diferentes tipos de vinho.

O objetivo principal do conjunto de dados é distinguir três variedades de vinho a partir de 13 variáveis químicas que medem compostos como o teor de álcool, ácido málico, magnésio, entre outros.

Tipos de Dados:

Características : 13 variáveis quantitativas que descrevem diferentes atributos químicos de cada amostra de vinho.

Rótulo (target): Um rótulo categórico que indica a classe do vinho (0, 1 ou 2), representando as três variedades de vinho.

alcohol: Teor alcoólico (%), uma medida do conteúdo de álcool em cada amostra.
malic_acid: Quantidade de ácido málico (g/l), que contribui para a acidez do vinho.

ash: Quantidade de cinzas (g/l), que são os resíduos inorgânicos após a queima do vinho.

alcalinity_of_ash: Alcalinidade das cinzas, uma medida do nível de álcalis nas cinzas.

magnesium: Concentração de magnésio (mg/l), um mineral essencial no vinho.
total_phenols: Quantidade total de fenóis, compostos que afetam o sabor e a cor do vinho.

flavanoids: Subconjunto de fenóis, que impacta diretamente no sabor do vinho e suas propriedades antioxidantes.

nonflavanoid_phenols: Fenóis que não são flavonoides, também afetam o sabor e a cor.

proanthocyanins: Taninos, que afetam a adstringência e o sabor do vinho.

color_intensity: Intensidade da cor do vinho, que pode influenciar a percepção de qualidade.

hue: Matiz da cor, uma medida de tonalidade.

od280/od315_of_diluted_wines: Relação entre absorções a 280 e 315 nm, usada para caracterizar compostos fenólicos.

proline: Quantidade de prolina, um aminoácido que pode indicar a maturidade da uva.

target: O rótulo ou classe do vinho, que pode ser 0, 1 ou 2, representando as três variedades de vinho.

```python
mean_values = wine_df.drop('target', axis=1).mean()

plt.figure(figsize=(10, 6))
mean_values.plot(kind='bar', color='skyblue')
plt.title('Médias das Variáveis Relevantes no Wine Dataset')
plt.ylabel('Média')
plt.xlabel('Variáveis')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

A correlação de Pearson é uma medida estatística que avalia a força e a direção da relação linear entre duas variáveis contínuas. O coeficiente de correlação de Pearson, representado por r, varia de -1 a 1:

r = 1: Correlação linear positiva perfeita (quando uma variável aumenta, a outra também aumenta).

r = -1: Correlação linear negativa perfeita (quando uma variável aumenta, a outra diminui).

r = 0: Sem correlação linear (não há uma relação linear entre as variáveis).

Correlação forte: Valores de r próximos de -1 ou 1.

Correlação fraca: Valores de r próximos de 0.

A correlação de Pearson é útil para entender a relação entre variáveis numéricas e pode orientar na escolha de variáveis em modelos preditivos ou análise exploratória de dados.
